```python
import requests 
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
```

# Cabinet of Malaysia

```python
urls = ['https://zh.wikipedia.org/zh-my/%E6%85%95%E5%B0%A4%E4%B8%81%E5%86%85%E9%98%81',
        'https://zh.wikipedia.org/zh-my/%E7%AC%AC%E4%B8%83%E6%AC%A1%E9%A9%AC%E5%93%88%E8%BF%AA%E5%86%85%E9%98%81',
        'https://zh.wikipedia.org/zh-my/%E7%AC%AC%E4%BA%8C%E6%AC%A1%E7%BA%B3%E5%90%89%E5%86%85%E9%98%81']

names = []
roles = []
cabinets = []

for ind, url in enumerate(urls):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    cabinet = soup.find('h1', id='firstHeading').text
    
    # Get the cabinet table 
    for table in soup.find_all('table', class_='sortable wikitable'):
        for row in table.find_all('tr'): 
            head = row.find('th') # get role in 1st column
            column = row.find_all('td') # get name in 3rd column  
            
            if head:
                role = ','.join([r.text for r in head.find_all('a')])
                if len(role) != 0:
                    roles.append(role)
                    #print(role)
            else:
                roles.append(role)
                
            if len(column) != 0:
                name = column[1].find_all('a')[0]
                names.append(name.text)
                cabinets.append(cabinet)
                
df = pd.DataFrame({'cabinet': cabinets, 'name': names, 'role': roles})
df.head()
```

```python
df.to_csv('cabinets.csv', index=False, encoding='utf-8-sig')
```

```python
df.name.to_csv(r'cabinets.txt', header=None, index=False)
```
